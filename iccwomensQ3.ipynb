```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://www.icc-cricket.com/rankings/womens/team-rankings/odi')
```

```python
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
teams = []

for i in soup.find_all('tr',class_="table-body"):
    teams.append((i.text.split('\n\n\n')[1]).split('\n')[0])
    
    
teams 
```

```python
matches = []

for i in soup.find_all('tr',class_="table-body"):
    matches.append(((i.text.split('\n\n\n')[1]).split('\n\n')[1]).split('\n')[0])
    
    
matches 
```

```python
points = []

for i in soup.find_all('tr',class_="table-body"):
    points.append(((i.text.split('\n\n\n')[1]).split('\n\n')[1]).split('\n')[1])
    
    
points 
```

```python
ratings = []

for i in soup.find_all('td',class_="table-body__cell u-text-right rating"):
    ratings.append(i.text)
    
    
ratings
```

```python
#Data frame
import pandas as pd
df = pd.DataFrame({'Teams':teams,'Matches' :matches,'Points' :points, 'Ratings' :ratings})
df.head(10)
```

```python
page1 = requests.get('https://www.icc-cricket.com/rankings/womens/player-rankings/odi/batting')
```

```python
page1
```

```python
soup1=BeautifulSoup(page1.content)
soup1
```

```python
players = []

for i in soup1.find_all('td',class_="table-body__cell rankings-table__name name"):
    players.append(i.text.split('\n')[1])
    
    
players
```

```python
team = []

for i in soup1.find_all('span',class_="table-body__logo-text"):
    team.append(i.text)
    
    
team 
```

```python
playerRating = []

for i in soup1.find_all('td',class_="table-body__cell rating"):
    playerRating.append(i.text)
    
    
playerRating
```

```python
#Data frame
import pandas as pd1
df1 = pd1.DataFrame({'Player':players,'Team' :team,'Rating' :playerRating})
df1.head(10)
```
