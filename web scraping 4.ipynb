```python
driver=webdriver.Chrome()
```

```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
time.sleep(5)
```

```python
rows=[]
ranks=[]
names=[]
artists=[]
uploadDate=[]
views=[]
search_Rank=driver.find_elements(By.XPATH,"//html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[1]/tbody/tr")
count=1;
for i in search_Rank:
    coulmns=driver.find_elements(By.XPATH,"//html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[1]/tbody/tr["+str(count)+"]/td")
    ranks.append(coulmns[0].text)
    names.append(coulmns[1].text)
    artists.append(coulmns[2].text)
    uploadDate.append(coulmns[4].text)
    views.append(coulmns[3].text)
    count=count+1  
uploadDate
```

```python
import pandas as pd
df=pd.DataFrame({'Rank':ranks,'Names':names,'Artist':artists,'Upload Date':uploadDate, 'Views':views})
df
```

```python
driver.get("https://www.bcci.tv/")
time.sleep(5)
```

```python
international=driver.find_element(By.XPATH,"//a[text()='INTERNATIONAL']")
international.click()
```

```python
seriesData=[]
datesData=[]
timesData=[]
placeData=[]
titlesData=[]
series=driver.find_elements(By.XPATH,'//h5[@class="match-tournament-name ng-binding"]')
for i in series:
    seriesData.append(i.text)
dates=driver.find_elements(By.XPATH,'//div[@class="match-dates ng-binding"]')
for i in dates:
    datesData.append(i.text)
times=driver.find_elements(By.XPATH,'//div[@class="match-time no-margin ng-binding"]')
for i in times:
    timesData.append(i.text)
place=driver.find_elements(By.XPATH,'//span[@class="ng-binding ng-scope"]')
for i in place:
    placeData.append(i.text)
titles=driver.find_elements(By.XPATH,'//span[@class="matchOrderText ng-binding ng-scope"]')
for i in titles:
    titlesData.append(i.text)
datesData
```

```python
import pandas as pd
df=pd.DataFrame({'Titles':titlesData,'Place':placeData,'Date':datesData,'Time':timesData, 'Series':seriesData})
df
```
