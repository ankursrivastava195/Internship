```python
#Ans-2
name=[]
price=[]
expectedDelivery=[]
Availability=[]
productUrl=[]
```

```python
count = 0;
while(count < 3):
    title_tags=driver.find_elements(By.XPATH,'//div[@class="a-section a-spacing-small puis-padding-left-small puis-padding-right-small"]')
    for i in title_tags:
        titlevalue=i.text
        if(titlevalue.split("\n")[0] == 'Sponsored'):
            name.append((titlevalue.split("\n")[1]))
        else:
            name.append((titlevalue.split("\n")[0]))
        try:
            price.append((titlevalue.split("\n₹")[1]).split(" M.R.P:")[0])
        except IndexError as e:
            price.append('-')
        try:
            expectedDelivery.append((titlevalue.split("Get it by")[1]).split("\n")[0])
            Availability.append("yes")
        except IndexError as e:
            expectedDelivery.append('-')
            Availability.append("no")

    url_tags=driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-no-outline"]')
    for i in url_tags[0:len(title_tags)]:
        url=i.get_attribute ('href')
        productUrl.append(url)
    if(count < 2):
        nextPage = driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]') 
        count = count + 1
        nextPage.click()
        time.sleep(5)
    else:
        break
```

```python
len(productUrl)
```

```python
import pandas as pd
df=pd.DataFrame({'Name':name,'Price':price,'ExpectedDelivery':expectedDelivery,'Availability':Availability, 'Product Url':productUrl})
df
```

```python
driver.get("https://flipkart.com")
```

```python
search_button=driver.find_element(By.CLASS_NAME,"_3704LK")
search_button.send_keys('Oneplus nord')
```

```python
search_click=driver.find_element(By.CLASS_NAME,"L0Z3Pu")
search_click.click()
```

```python
BrandName=[]
Product_smartphoneName=[]
Product_Color=[]
Product_RAM=[]
Product_ROMStorage=[]
Product_Primarycamera=[]
Product_SecondaryCamera=[]
Product_DisplaySize=[]
Product_BatteryCapacity=[]
Product_Price=[]
Product_ProductURL=[]
```

```python
Product_tags=driver.find_elements(By.XPATH,'//div[@class="_2kHMtA"]')
for i in Product_tags:
    product=i.text
    if(product.split('\n')[0] == 'Bestseller' or product.split('\n')[0] == "Flipkart's Choice" or product.split('\n')[0] == 'Currently unavailable'):
        Product_BrandName = product.split('\n')[2]
    else:
        Product_BrandName = product.split('\n')[1]
    BrandName.append(Product_BrandName)
    try:
        Product_Color.append((Product_BrandName.split('(')[1]).split(',')[0])
    except IndexError as e:
        Product_Color.append('-')
    try:
        Product_RAM.append((product.split('Reviews\n')[1]).split('RAM')[0])
    except IndexError as e:
        Product_RAM.append('-')
    Product_ROMStorage.append((product.split('| ')[1]).split('ROM')[0])
    Product_DisplaySize.append((product.split('ROM\n')[1]).split('Display')[0])
    Product_BatteryCapacity.append((product.split('Camera\n')[1]).split('Battery')[0])
    try:
        Product_Primarycamera.append((product.split('Camera | ')[1]).split('Front Camera\n')[0])
    except IndexError as e:
        Product_Primarycamera.append('-')
    try:
        if('+' in (product.split('Display\n')[1]).split('Rear Camera')[0]):
            Product_SecondaryCamera.append(((product.split('Display\n')[1]).split('Rear Camera')[0]).split('|')[0])
        else:
            Product_SecondaryCamera.append((product.split('Display\n')[1]).split('Rear Camera')[0])
    except IndexError as e:
        Product_SecondaryCamera.append('-')
Price_tags=driver.find_elements(By.XPATH,'//div[@class="_30jeq3 _1_WHN1"]')
for i in Price_tags:
    price=i.text
    Product_Price.append(price)
    
Product_Price 

Url_tags=driver.find_elements(By.XPATH,'//a[@class="_1fQZEK"]')
for i in Url_tags[0:24]:
    url=i.get_attribute ('href')
    Product_ProductURL.append(url)
    
len(Product_smartphoneName)
```
